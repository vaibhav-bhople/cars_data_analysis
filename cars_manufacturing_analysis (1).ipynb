```python
import pandas as pd
```

```python
df = pd.read_csv("Cars_manufacturing_data.csv")
```

```python
df.head()
```

```python
del df['Unnamed: 0']
del df['vin']
del df['condition']
```

```python
df.head()
```

##To get list of unique countries

```python
df['brand'].unique()
```

```python
df[df['country'] == ' canada']
```

##**Top 5 car brands according to no. of cars**

```python
data = []

for country in df['brand'].unique():
  data.append([country, len(df[df['brand']==country])])

data = pd.DataFrame(data,columns=['brand','frequency'])
data.sort_values(by= 'frequency', ascending=False)
```

```python
data.sort_values(by= 'frequency', ascending=False).head()
```

```python
data.plot(x='brand',y='frequency',kind='bar')
```

##**Number of cars sold per year**

```python
data = []

for i in df['year'].unique():
  data.append([i, len(df[df['year']==i])])

data = pd.DataFrame(data,columns=['year','frequency'])
data.sort_values(by= 'frequency', ascending=True)
```

```python
data.plot(x='year',y='frequency',kind='bar')
```

##**Most popular car colors**

```python
data = []

for i in df['color'].unique():
  data.append([i, len(df[df['color']==i])])

data = pd.DataFrame(data,columns=['color','frequency'])
data.sort_values(by= 'frequency', ascending=False).head()
```

```python
data.plot(x='color',y='frequency',kind='bar',figsize=(20,5))
```

##**Max average price of cars**

```python
price_df=df.groupby('brand').mean()['price'].sort_values(ascending=False)
price_df
```

```python
price_df.plot(x='brand',y='price',kind='bar')
```
